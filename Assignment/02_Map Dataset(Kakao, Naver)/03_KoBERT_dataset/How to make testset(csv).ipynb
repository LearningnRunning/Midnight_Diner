## 전처리(중복값, 결측값 drop)

```python
tmp2 = tmp2.dropna(axis=0)
tmp2 = tmp2.drop_duplicates(['review','reviews_date','reviews_id'], keep='first')
limit_review_num = 100 #음식점 리뷰 개수
region_df= tmp2[tmp2['review_num'] >= limit_review_num].loc[:,['name','review']]
region_df
```

```python
test = pd.DataFrame()
set1 = set(region_df['name'].tolist())
res_count = 100
review_count = 50

for res in list(set1)[6:res_count+6]:
    print(res)
    region_df= tmp2.loc[:,['name','review']] # 컬럼 2개만 축출
    tmp = region_df[region_df.loc[:,'name']== res] # 특정 음식점
    first_index = region_df.index[region_df.loc[:,'name']== res][0] #인덱스 
    tmp = tmp.loc[first_index:first_index+review_count,:]
    test = pd.concat([test, tmp.loc[first_index:first_index+review_count,:]])
# 데이터 프레임의 최대 개수를 없애줍니다.
test = test.reset_index(drop=True)
# master.csv라는 파일로 만들어줍니다.
tmp.to_csv('test{0}_{1}.csv'.format(res_count,review_count), index=False)
```
