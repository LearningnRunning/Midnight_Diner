```python
okt_tokenizer = []
```

```python
for i in jangdari['pyko_review']:
    okt_tokenizer.append(okt.pos(i))
```

```python
okt_tokenizer_del_stwds = []
```

```python
for i in okt_tokenizer:
    token_words = []
    for j in i:
        if ('Josa' not in j) and ('Punctuation' not in j) and ('Foreign' not in j) and ('KoreanParticle' not in j):
            token_words.append(j)
    okt_tokenizer_del_stwds.append(token_words)
```
