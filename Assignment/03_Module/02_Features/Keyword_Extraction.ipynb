```python
def f(t, d):
    return d.count(t)

def tf(t, d):
    return 0.5 + 0.5 * f(t, d) / max([f(w, d) for w in d])

from math import log10

def idf(t, D):
    # D is documents == document list
    num = len(D)
    den = 1 + len([True for d in D if t in d])
    return log10(num / den)

def tfidf(t, d, D):
    return tf(t,d)*idf(t, D)

def tfidfScorer(D):
    result = []
    for d in D:
        result.append([(t, tfidf(t, d, D)) for t in d])
    return result
```

```python
import numpy as np
import itertools

from konlpy.tag import Okt
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from sentence_transformers import SentenceTransformer
```

```python
df = pd.read_csv('D:/Midnight_Diner/Assignment/02_Map Dataset(Kakao, Naver)/03_KoBERT_dataset/강남역/test10_5.csv')
df.head()
```

```python
okt = Okt()

tokenized_doc = df["review"].apply(okt.pos)
df.head()
```

```python
df["tok1"] = df["review"].apply(okt.morphs)
```

```python
df['tok1'][1]
```

```python
df['tokenized_doc'] = df["review"].apply(okt.pos)
df.head()
```

```python
df['tokenized_nouns'] = df['tokenized_doc'].apply(lambda x :(' '.join([word[0] for word in x if word[1] == 'Noun'])))
df.head()
```

```python
n_gram_range = (1, 2)

count = CountVectorizer(ngram_range=n_gram_range).fit(df['tokenized_nouns'])
candidates = count.get_feature_names_out()

print('Bigram 개수 :',len(candidates))
print('Bigram 다섯개만 출력 :',candidates[:5])
```

```python
model = SentenceTransformer('sentence-transformers/xlm-r-100langs-bert-base-nli-stsb-mean-tokens')
df['doc_embedding'] = df['review'].apply(lambda x: model.encode(x))
df['candidate_embeddings'] = df['review'].apply(lambda x: model.encode(x))
```

```python
df['for_cosine_sim'] = df['review'].apply(
    lambda x: ((model.encode(x), model.encode(x)) for (doc_embedding, candidate_embeddings) in x))
```

```python
df.head()
```

```python
# top_n = 10
df['distance'] = list(cosine_similarity(list(df['doc_embedding'].values), list(df['candidate_embeddings'].values)))
```

```python
top_n = 10
df['keywords'] = df['distance'].apply(lambda x: [candidates[index] for index in x.argsort()[-top_n:]])
df['keywords'].head()
```
